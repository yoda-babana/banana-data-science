## 1.6.  Joining, Merging and Concatenation of DataFrames

```python
dFrame1=pd.DataFrame([[1, 2, 3], [4, 5],[6]], columns=['C1', 'C2', 'C3'], index=['R1','R2', 'R3'])
dFrame1
```

```python
dFrame2=pd.DataFrame([[10, 20], [30], [40,50]], columns=['C2', 'C5'], index=['R4', 'R2','R5'])
dFrame2
```

```python
dFrame1=dFrame1._append(dFrame2)
dFrame1
```

```python
# append dFrame1 to dFrame2
dFrame2 =dFrame2._append(dFrame1,sort=True)
dFrame2
```

```python
# append dFrame1 to dFrame2 with sort=False
dFrame2 = dFrame2._append(dFrame1,sort=False)
dFrame2
```

```python
dFrame1 = dFrame1._append(dFrame2, ignore_index=True)
dFrame1
```
